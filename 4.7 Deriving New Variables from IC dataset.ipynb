# 1. User-defined function on subset df

```python
# Creating subset
df = ords_prods_merge[:1000000]
```

```python
# Checking shape
df.shape
```

```python
# Defining function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Checking max

df['prices'].max()
```

# 2. Creating price_label with .loc() function

```python
ords_prods_merge.loc[ords_prods_merge['prices'] > 15, 'price_range_loc'] = "High-range product"
```

```python
ords_prods_merge.loc[(ords_prods_merge['prices'] <= 15) & (ords_prods_merge['prices'] > 5), 'price_range_loc'] = "Mid-range product"
```

# 3. Creating 'busiest_day' column with for-loop

```python
# Frequency of 'orders_day_of_the_week'
ords_prods_merge['orders_day_of_the_week'].value_counts(dropna = False)
```

### 'busiest_day' = 0, 'least_busy' = 4

```python
# Creating empty list for values
result = []

# for-loop statement to categorize values 
for value in ords_prods_merge["orders_day_of_the_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# Calling results from for-loop
result
```

```python
# Setting results of for-loop into new column 
ords_prods_merge['busiest_day'] = result
```

```python
# Frequency of 'busiest_day' column
ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

# 4. Creating new columns: 'busiest_days' & 'slowest_days'

```python
# Dropping 'busiest_day' column
ords_prods_merge.drop('busiest_day', axis=1, inplace=True)
```

```python
# Renaming df after dropping 'busiest_day' column
ords_prods_new = ords_prods_merge
```

```python
# Frequency of 'orders_day_of_the_week'
ords_prods_new['orders_day_of_the_week'].value_counts(dropna = False)
```

```python
# "Busiest days" are 0 (Sat) and 1 (Sun)

# Creating busiest days column with .loc() function
ords_prods_new.loc[(ords_prods_new['orders_day_of_the_week'] == 0) | (ords_prods_new['orders_day_of_the_week'] == 1), 'busiest_days'] = 'Busiest days'
```

```python
# "Slowest days" are 4 (Wed) and 3 (Tues)

# Creating slowest days column with .loc() function
ords_prods_new.loc[(ords_prods_new['orders_day_of_the_week'] == 4)| (ords_prods_new['orders_day_of_the_week'] == 3), 'slowest_days'] = 'Slowest days'
```

```python
# *(Sat + Sun) = (6204182 + 5660230) = 11864412

# Checking if counts match frequency 'busiest_day'
ords_prods_new['busiest_days'].value_counts(dropna = False)
```

```python
# *(Tues + Wed)= (3840534 + 3783802) = 7624336

# Checking if counts match frequency 'busiest_day'
ords_prods_new['slowest_days'].value_counts(dropna = False)
```

# 5. Creating new column: 'busiest_period_of_day'

```python
# Checking frequency of 'order_hour_of_day'
ords_prods_new['order_hour_of_day'].value_counts(dropna = False)
```

### Insights: Hours with "Most orders" = 9,10,11,12,13,14,15,16,17;  Hours with "Average orders" = 6,7,8,18,19,20,21,22,23; Hours with "Least orders" = 0,1,2,3,4,5

```python
# Simplifying time-frames into lists
list_1 = [9, 10, 11, 12, 13, 14, 15, 16, 17]
list_2 = [0, 1, 2, 3, 4, 5]

# Creating empty list for values
results = []

# for-loop statement to categorize values
for value in ords_prods_new["order_hour_of_day"]:
    if value in (list_1):
        results.append("Most orders")
    elif value in (list_2):
        results.append("Fewest orders")
    else:
        results.append("Average orders")
```

```python
# Calling results
results
```

```python
# Setting results of for-loop into new column 
ords_prods_merge['busiest_period_of_day'] = result
```

```python
# Checking frequency of 'busiest_period_of_the_day' column
ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
